```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
july_2 = pd.read_csv("/content/drive/My Drive/Insight/Data/july2_merged.csv")
july_2.head()
```

```python
two_days = pd.concat([july_2, merged])
```

```python
two_days.describe()
```

```python
overall_rms = sqrt(mean_squared_error(two_days.car_count, two_days.yolo_car_count))
overall_rms
```

```python
two_days.groupby("Parking_lot").apply(
    lambda x: mean_squared_error(x.car_count, x.yolo_car_count) ** .5)
```

```python
two_days["date_time"] = two_days.apply(lambda x: "{:02d}/{:02d}/{:4d}/{:02d}/{:02d}".format(x.month,x.day,x.year,x.hour,x.minute), axis= 1)
```

```python
format = '%m/%d/%Y/%H/%M'
two_days['date_time'] = pd.to_datetime(two_days['date_time'], format = format)
two_days.info()
```

```python
from plotnine import *
```

```python
ggplot(two_days, aes(x = "car_count", y = "yolo_car_count", color = "Parking_lot")) +\
  geom_point(alpha = .7) +\
  geom_abline( slope = 1, intercept = 0) +\
  coord_cartesian(xlim = (0,15), ylim = (0,15)) +\
  labs(x = "Actual Car Count", y = "Model (yolo) Car Count") +\
  theme_bw() +\
  facet_wrap("Parking_lot")
```

```python
ggplot(two_days, aes(x = "date_time")) +\
  geom_line(aes(y = "car_count"), color = "green") +\
  geom_line(aes(y = "yolo_car_count"), color = "blue") +\
  labs(x = "Time", y = "Car Count") +\
  theme_bw() +\
  facet_wrap("Parking_lot")
```
